```python
for n in range (1,101):
    if (n % 5 == 0) and (n % 2 != 0):
        print (str(n) + ' is odd and high five!')
    elif (n % 2 != 0):
        print (str(n) + ' is odd')
    elif (n % 2 == 0) and (n % 7 == 0):
        print (str(n) + ' is divisible by 14')
    elif (30 < n < 45) and (n % 8 == 0):
        print (str(n)+' this happens twice!')
    else:
        print (n)
```
